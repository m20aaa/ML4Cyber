Функция get_config('X') возвращает набор признаков после предобработки.

Как можно увидеть, численные значения признаков были нормализованы с помощью Z-оценки, а категориальные призанки были закодированы с помощью метода быстрого кодирования ("one-hot encoding"). Очень важно проверить результаты нормализации, в некоторых случаях она может быть выполнена некорректно, что повлияет на возможность обучить классификатр и получить верные занчения.

Нормализация Z-оценки относится к процессу нормализации каждого значения в наборе данных, так что среднее значение всех значений равно 0, а стандартное отклонение равно 1.

Мы используем следующую формулу для нормализации z-оценки для каждого значения в наборе данных:

Новое значение = (x – μ) / σ

где:

х : Исходное значение

μ : Среднее значение данных

σ : стандартное отклонение данных
